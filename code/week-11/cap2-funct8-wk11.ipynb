```python
# The Sensitivity Check: Which Variable is the "Trig point"?
import pandas as pd

def get_rf_sensitivity(x_point, model, epsilon=0.01):
    sensitivities = []
    base_pred = model.predict(x_point.reshape(1, -1))[0]
    
    for i in range(8):
        # Nudge up
        x_plus = x_point.copy()
        x_plus[i] = np.clip(x_plus[i] + epsilon, 0, 1)
        pred_plus = model.predict(x_plus.reshape(1, -1))[0]
        
        # Nudge down
        x_minus = x_point.copy()
        x_minus[i] = np.clip(x_minus[i] - epsilon, 0, 1)
        pred_minus = model.predict(x_minus.reshape(1, -1))[0]
        
        # Calculate local gradient (absolute change)
        grad = (abs(pred_plus - base_pred) + abs(pred_minus - base_pred)) / (2 * epsilon)
        sensitivities.append(grad)
        
    return np.array(sensitivities)

# Run sensitivity on your Recommended Point
recommended_x = np.array([0.177523, 0.106972, 0.139817, 0.0, 1.0, 0.479984, 0.294613, 0.100080])
sens = get_rf_sensitivity(recommended_x, model)

# Display Results
df_sens = pd.DataFrame({'Dimension': range(8), 'Sensitivity': sens})
print("\n--- Local Sensitivity at Recommended Point ---")
print(df_sens.sort_values(by='Sensitivity', ascending=False).to_string(index=False))
```

```python
# 1. Setup Data
X_arr = np.array(X)
Y_arr = np.array(Y).flatten()

# 2. Train Random Forest (The "Sharp" Surrogate)
# RF is better at "step functions" than MLP
model = RandomForestRegressor(n_estimators=100, min_samples_split=2, random_state=42)
model.fit(X_arr, Y_arr)

# 3. Define Acquisition Function (UCB)
# This rewards predictions that are high, but also rewards high variance (uncertainty)
def ucb_acquisition(x, model, kappa=1.5):
    # Get predictions from all 100 trees to estimate uncertainty (sigma)
    preds = np.array([tree.predict(x.reshape(1, -1))[0] for tree in model.estimators_])
    mu = np.mean(preds)
    sigma = np.std(preds)
    
    # UCB score: The "optimistic" prediction
    return -(mu + kappa * sigma)

# 4. Multi-start Optimization
best_x = None
best_val = np.inf

print("Optimizing with UCB acquisition...")
for _ in range(50):
    x0 = np.random.uniform(0, 1, 8)
    res = minimize(ucb_acquisition, x0, args=(model,), bounds=[(0, 1)]*8, method='L-BFGS-B')
    
    if res.fun < best_val:
        best_val = res.fun
        best_x = res.x

print(f"Recommended Submission: {'-'.join(f'{v:.6f}' for v in best_x)}")

tree_predictions = np.array([tree.predict(best_x.reshape(1, -1))[0] for tree in model.estimators_])
predicted_y = np.mean(tree_predictions)
model_uncertainty = np.std(tree_predictions)
print(f"Predicted Y (Mean): {predicted_y:.4f}    Model Uncertainty: {model_uncertainty:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from scipy.optimize import minimize
from scipy.spatial.distance import cdist

# 1. DATA PREPARATION
# X and Y are assumed to be your lists/arrays as provided
X_arr = np.array(X)
Y_arr = np.array(Y).reshape(-1, 1)

# Scalers are essential for Neural Networks to treat all 8 dims equally
scaler_y = StandardScaler()
Y_scaled = scaler_y.fit_transform(Y_arr).ravel()

# 2. TRAIN THE NEURAL NETWORK
# We use 'tanh' for smooth gradients and 'lbfgs' for small-sample precision
nn = MLPRegressor(
    hidden_layer_sizes=(64, 32),
    activation='tanh',
    solver='lbfgs',
    alpha=0.5,           # High regularization to smooth the "9.19 cliff"
    max_iter=5000,
    random_state=42
)
nn.fit(X_arr, Y_scaled)

# 3. DEFINE ROBUST OBJECTIVE (The "Safety Rope" Search)
best_idx = np.argmax(Y_arr)
x_best_ever = X_arr[best_idx]

def robust_objective(x):
    x_input = x.reshape(1, -1)
    
    # NN Predicted Score
    pred_scaled = nn.predict(x_input)[0]
    
    # Distance Penalty (Prevents sliding into "fake" corner peaks)
    # We penalize the Euclidean distance from our best known point
    distance = cdist(x_input, x_best_ever.reshape(1, -1), metric='euclidean')[0][0]
    
    # LAMBDA (0.3) controls how much we trust the NN vs. staying near data
    # Increase to 1.0 to stay very close to 9.899, decrease to 0.1 to explore more.
    penalty_weight = 0.3 
    return -pred_scaled + (penalty_weight * distance)

# 4. RUN CONSTRAINED OPTIMIZATION
# We start at the 9.899 point and look for a local improvement
res = minimize(
    robust_objective, 
    x0=x_best_ever, 
    bounds=[(0, 1)] * 8, 
    method='L-BFGS-B'
)

# 5. POST-PROCESS RESULTS
final_x = res.x
final_y_pred_scaled = nn.predict(final_x.reshape(1, -1))
final_y_pred = scaler_y.inverse_transform(final_y_pred_scaled.reshape(-1, 1))[0][0]

# 6. OUTPUTS
print("--- ROBUST NN OPTIMIZATION ---")
print(f"Current Best Score in Data: {Y_arr[best_idx][0]:.4f}")
print(f"NN Predicted Y at New Point: {final_y_pred:.4f}")
print("\nSuggested Submission (X0-X7):")
print('-'.join(f'{v:.6f}' for v in final_x))

# 7. FEATURE IMPORTANCE (Sensitivity at this specific point)
def get_sensitivity(x):
    eps = 1e-5
    grads = []
    base = nn.predict(x.reshape(1, -1))
    for i in range(len(x)):
        x_p = x.copy()
        x_p[i] += eps
        grads.append(abs(nn.predict(x_p.reshape(1, -1)) - base) / eps)
    return np.array(grads).flatten()

sens = get_sensitivity(final_x)
importance_df = pd.DataFrame({'Dim': range(8), 'Sensitivity': sens})
print("\nLocal Sensitivity (Which variables are pushing the score up here?):")
print(importance_df.sort_values(by='Sensitivity', ascending=False).to_string(index=False))
```

```python
# 1. SETUP & DATA
X = np.array(X) # Loaded from your data
Y = np.array(Y) # Loaded from your data
n_dims = X.shape[1]
y_max = np.max(Y)

# 2. IMPROVED KERNEL
# Increased n_restarts and adjusted length_scale bounds for [0, 1] space
kernel = C(1.0, (1e-2, 1e3)) * Matern(
    length_scale=[0.5] * n_dims, 
    length_scale_bounds=(1e-2, 1e2), # 100 is plenty for a 0-1 range
    nu=2.5
) + WhiteKernel(
    noise_level=1e-5, 
    noise_level_bounds=(1e-12, 1e-2)   )

model = GaussianProcessRegressor(
    kernel=kernel,
    alpha=1e-6, # Slightly higher alpha for numerical stability
    normalize_y=True,
    n_restarts_optimizer=25 # Sufficient restarts for 8D
)
print("Fitting model...")
model.fit(X, Y)
print(f"Model Kernel: {model.kernel_}")
```

```python
# 3. VECTORIZED EI FUNCTION
def get_ei(x_points, model, y_max):
    x_points = np.atleast_2d(x_points)
    mu, sigma = model.predict(x_points, return_std=True)
    sigma = sigma.reshape(-1, 1)
    mu = mu.reshape(-1, 1)
    with np.errstate(divide='warn'):
        imp = mu - y_max
        Z = imp / sigma
        ei = imp * norm.cdf(Z) + sigma * norm.pdf(Z)
        ei[sigma <= 0.0] = 0.0
    return ei.flatten()
```

```python
# 4. CANDIDATE GENERATION (Large Scale)
num_candidates = 250000 
x_candidates = np.random.uniform(0, 1, size=(num_candidates, n_dims))
# Include current best points and their slight perturbations
best_indices = np.argsort(Y)[-5:]
x_candidates = np.concatenate([x_candidates, X[best_indices]], axis=0)

ei_vals = get_ei(x_candidates, model, y_max)
# Get top 5 starting points for multi-start optimization
top_idx = np.argsort(ei_vals)[-5:]
starts = x_candidates[top_idx]

# 5. MULTI-START POLISHER
best_res_x = None
best_res_ei = -np.inf

def objective(x):
    return -get_ei(x, model, y_max)[0]

for x0 in starts:
    res = minimize(objective, x0=x0, bounds=[(0, 1)] * n_dims, method='L-BFGS-B')
    if -res.fun > best_res_ei:
        best_res_ei = -res.fun
        best_res_x = res.x

x_final = np.clip(best_res_x, 0, 1)

# 6. RESULTS
print(f"Weekly Submission: {[f'{v:.8f}' for v in x_final]}")
print(f"Submit this: {'-'.join(f'{v:.6f}' for v in x_final)}")
print(f"Predicted Y: {model.predict(x_final.reshape(1, -1))[0]:.6f}")
```
